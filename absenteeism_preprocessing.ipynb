### Group the reason for absence

```python
#drop the reason for absence columns
data_with_no_id = data_with_no_id.drop('Reason for Absence', axis=1)
data_with_no_id
```

```python
#group the reason columns into 4 groups

group_reason_1 = reason_columns.loc[:,1:14]
group_reason_1
group_reason_2 = reason_columns.loc[:,15:17]
group_reason_2
group_reason_3 = reason_columns.loc[:,18:21]
group_reason_3
group_reason_4 = reason_columns.loc[:,22:28]
group_reason_4
```

```python
#get the max value at each row

reason_1 = group_reason_1.max(axis=1)
reason_1
reason_2 = group_reason_2.max(axis=1)
reason_2
reason_3 = group_reason_3.max(axis=1)
reason_3
reason_4 = group_reason_4.max(axis=1)
reason_4
```

```python
#Include the 4 groups of reasons into the dataset

data_with_no_id['Reason_1'] = reason_1
data_with_no_id['Reason_2'] = reason_2
data_with_no_id['Reason_3'] = reason_3
data_with_no_id['Reason_4'] = reason_4
data_with_no_id
```

```python
columns_names_reorder = ['Reason_1','Reason_2', 'Reason_3', 'Reason_4','Date', 'Transportation Expense',
                 'Distance to Work', 'Age','Daily Work Load Average', 'Body Mass Index', 'Education',
             'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
#Reorder the columns names by placing the reasons at the begining
data_with_no_id = data_with_no_id[columns_names_reorder]
data_with_no_id
```

### Create a Check point

```python
df_reason_mod = data_with_no_id.copy()
df_reason_mod
```

### Extract the month value and the day of the week from the ‘Date’ column

```python
#check the type of the column Date
type(df_reason_mod.Date[0])
```

```python
#convert the Date columns into date type
df_reason_mod.Date = pd.to_datetime(df_reason_mod.Date, format='%d/%m/%Y')
```

```python
#check the type of the Date column
type(df_reason_mod.Date[0])
```

```python
df_reason_mod.Date
```

```python
#df_reason_mod.Date[45].day

#create a list that will contain the day and month values

day_list =[]
for i in df_reason_mod.Date:
    day_list.append(i.day)
day_list

month_list =[]
for i in df_reason_mod.Date:
    month_list.append(i.month)
len(month_list)
```

```python
df_reason_mod['Month Value'] = month_list
df_reason_mod['Day of the Week'] = day_list

df_reason_mod
```

```python
#drop the ‘Date’ column

#df_reason_mod = df_reason_mod.drop(['Month','Day'], axis=1)
df_reason_mod = df_reason_mod.drop(['Date'], axis=1)
df_reason_mod
```

```python
df_reason_mod.columns.values
```

```python
columns_names_reorder_2 = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value',
       'Day of the Week','Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df_reason_mod = df_reason_mod[columns_names_reorder_2]
df_reason_mod
```

```python
#get the number of values in the columns Education
#1 for high school, 2 for graduate, 3 for postgraduate, 4 for master or doctor

df_reason_mod.Education.unique()
```

```python
#get the number of values in each category 
df_reason_mod.Education.value_counts()
```

```python
#Map 0 for a high school and 1 to the rest
df_reason_mod.Education = df_reason_mod.Education.map({1:0, 2:1, 3:1, 4:1})
df_reason_mod
```

```python
#get the new values after mapping
df_reason_mod.Education.unique()
```

```python
#get the new values after mapping
df_reason_mod.Education.value_counts()
```
